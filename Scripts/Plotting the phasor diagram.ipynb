```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import MonteCarlo as MC
import math 
```

```python
def universal_semicircle():
    gs = np.linspace(0,1,50)
    Ess = np.sqrt(0.25-(gs-0.5)**2)
    return gs, Ess
path = "/Volumes/SamuelSSD/School/KurfMonteCarlo/Program/Figures/"
date = "0110"
def Distance(s1,g1,s2,g2):
    dx = s2-s1
    dy = g2-g1
    dis = math.sqrt(dy**2+dx**2)
    return dis
```
